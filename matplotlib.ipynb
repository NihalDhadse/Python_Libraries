```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("C:/Users/nihal/OneDrive/Desktop/jupyter notebook/ElectricCarData_Clean.csv")
df=pd.DataFrame(data.head(50))

group=df.groupby("Brand").agg({"Range_Km":"sum"})
print(group)
plt.step(group.index,group["Range_Km"],marker="o")
plt.xticks(rotation=90) 
plt.show()

```

```python
#legend
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[22,35,30,13,10] 
y1=[54,24,61,47,16]
plt.figure(figsize=[4,3])                       
plt.plot(x,y,label=["male"])
plt.plot(x,y1,label=["female"])
plt.legend( loc=0,labelspacing=2,bbox_to_anchor=(1.4,0.7))                            
plt.savefig("mafemale.png",pad_inches=1.5,bbox_inches="tight")                     
plt.show()
```
