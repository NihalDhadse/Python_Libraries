```python
#bar plot
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data=sns.load_dataset("tips")                      
print(data)
sns.barplot(data=data, x="day", y="tip",estimator='median',hue="sex",palette="Blues",order=["Sun","Sat","Fri","Thur"])    
plt.plot()                                                
```

```python
#histogram plot
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data=sns.load_dataset("tips") 
sns.histplot(data,x="day",hue="sex",kde=True)
plt.show()
```

```python
#scatter plot
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data=sns.load_dataset("tips") 
sns.scatterplot(data=data,x="total_bill",y="tip",hue="smoker",palette="viridis")       
plt.legend()
plt.show()
```

```python
#heatmap 
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data=sns.load_dataset("tips") 
gp=data.groupby("day").agg({"tip":"mean"})
sns.heatmap(gp)
plt.show()
```

```python
#multiple plots in seaborn
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data=sns.load_dataset("tips")
a=sns.FacetGrid(data,col="smoker",height=3)    
a.map(sns.barplot,"day","tip")
plt.show()
```

```python
#multiple plots in seaborn
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data=sns.load_dataset("tips")
a=sns.FacetGrid(data,col="time",height=3,hue="sex")   
a.map(sns.barplot,"day","tip")
plt.show()
```

```python
#relational plot
import seaborn as sns
import matplotlib.pyplot as plt
data=sns.load_dataset("tips")
sns.relplot(data,x="tip",y="total_bill",hue="sex",col="day")
plt.show()
```

```python
#swarm plot
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data=sns.load_dataset("tips")
sns.swarmplot(data,x="day",y="total_bill",hue="sex")     
plt.show()
```

```python
#kde plot (kernal density estimation )
import seaborn as sns
import matplotlib.pyplot as plt
data=sns.load_dataset("tips")
sns.kdeplot(data,x="total_bill",hue="day",multiple="stack")       
plt.show()
```
